```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#datayı import edelim
df = pd.read_csv("/Users/haticecakir/Downloads/amazon.csv")
df
```

```python
#veri türlerine, sütunlara, satır numarasına bir göz atalım
df.info()
```

```python
df['actual_price'] #para birimi işaretini ve virgülü kaldırmamız gerekiyor, böylece float'a dönüştürebileceğiz
df['discounted_price']
```

```python
df.info()
```

```python
df['discount_percentage']  # % işaretini kaldırmamız gerekiyor, böylece float'a dönüştürebileceğiz

df['discount_percentage']= df['discount_percentage'].replace( {'%': ''}, regex=True).astype(float)/100
df.info()
```

```python
#derecelendirme sütunu, veri türünü değiştirmemiz gerekiyor
df['rating'] #bu sütundaki tüm benzersiz değerlere bir göz atalım

sorted(df['rating'].unique(), reverse=False)
#rating 2 (min) ile 5 (maks) arasındadır ve orada | kaldırılması veya değiştirilmesi gereken şey.
```

```python
#\'li satırı derecelendirme olarak görelim
df[df['rating']== '|']
#bu ürün için başka incelememiz var mı bakalım
#Veri kümem küçük olduğundan, yalnızca satırı kaldırmak yerine onu başka bir şeyle değiştirip değiştiremeyeceğimi görelim.
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce') #coerce hata almak yerine sayı olmayanı NA olarak değiştirmeye yardımcı olur

#yani artık NA olarak buna sahibim ve onu Mean ile değiştirmem gerekiyor

avg_rating = df['rating'].mean()

df['rating'] = df['rating'].fillna(avg_rating)
```

```python
#boş derecelendirmeli satırları gösterelim, ayrıca format artık float.
df[df['rating'].isnull()]
```

```python
df.info()
```

```python
#rating_count'a bir göz atalım, 2 tane null olduğunu zaten biliyoruz
df[df['rating_count'].isnull()]
```

```python
#bu 2 ürünü B0B94JPY2N, B0BQRJ3C47 için kopyalayıp kopyalamadığımızı görelim
#df[df['product_id']=='B0BQRJ3C47']

df[df['product_id'].isin(['B0B94JPY2N', 'B0BQRJ3C47'])]
#hayır! bu satırları silebilir veya verilerde en az bir inceleme olduğundan NA'yı 1 ile değiştirebilir veya bunları avg(rating_count) ile değiştirebiliriz.
```
