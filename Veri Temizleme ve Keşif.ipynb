```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#datayƒ± import edelim
df = pd.read_csv("/Users/haticecakir/Downloads/amazon.csv")
df
```

```python
#veri t√ºrlerine, s√ºtunlara, satƒ±r numarasƒ±na bir g√∂z atalƒ±m
df.info()
```

```python
df['actual_price'] #para birimi i≈üaretini ve virg√ºl√º kaldƒ±rmamƒ±z gerekiyor, b√∂ylece float'a d√∂n√º≈üt√ºrebileceƒüiz
df['discounted_price']
```

```python
df.info()
```

```python
df['discount_percentage']  # % i≈üaretini kaldƒ±rmamƒ±z gerekiyor, b√∂ylece float'a d√∂n√º≈üt√ºrebileceƒüiz

df['discount_percentage']= df['discount_percentage'].replace( {'%': ''}, regex=True).astype(float)/100
df.info()
```

```python
#derecelendirme s√ºtunu, veri t√ºr√ºn√º deƒüi≈ütirmemiz gerekiyor
df['rating'] #bu s√ºtundaki t√ºm benzersiz deƒüerlere bir g√∂z atalƒ±m

sorted(df['rating'].unique(), reverse=False)
#rating 2 (min) ile 5 (maks) arasƒ±ndadƒ±r ve orada | kaldƒ±rƒ±lmasƒ± veya deƒüi≈ütirilmesi gereken ≈üey.
```

```python
#\'li satƒ±rƒ± derecelendirme olarak g√∂relim
df[df['rating']== '|']
#bu √ºr√ºn i√ßin ba≈üka incelememiz var mƒ± bakalƒ±m
#Veri k√ºmem k√º√ß√ºk olduƒüundan, yalnƒ±zca satƒ±rƒ± kaldƒ±rmak yerine onu ba≈üka bir ≈üeyle deƒüi≈ütirip deƒüi≈ütiremeyeceƒüimi g√∂relim.
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce') #coerce hata almak yerine sayƒ± olmayanƒ± NA olarak deƒüi≈ütirmeye yardƒ±mcƒ± olur

#yani artƒ±k NA olarak buna sahibim ve onu Mean ile deƒüi≈ütirmem gerekiyor

avg_rating = df['rating'].mean()

df['rating'] = df['rating'].fillna(avg_rating)
```

```python
#bo≈ü derecelendirmeli satƒ±rlarƒ± g√∂sterelim, ayrƒ±ca format artƒ±k float.
df[df['rating'].isnull()]
```
